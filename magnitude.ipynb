```
import numpy as np
import pandas as pd
```

```
magnitude_data = pd.read_csv("magnitude.csv")
```

```
magnitude_data = magnitude_data[['Latitude', 'Longitude', 'Magnitude']]
magnitude_data.head()
```

```
magnitude_data = pd.get_dummies(magnitude_data)
magnitude_data.head()
```

```
from sklearn.model_selection import train_test_split

X=magnitude_data.drop('Magnitude', axis=1) 
y=magnitude_data.Magnitude  

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) 
```

```
X_train.head()
y_train.head()
```

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```
y_pred = regressor.predict(X_test)
```

```
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```
from joblib import dump, load
dump(regressor, 'magnitude_model.joblib') 
```
